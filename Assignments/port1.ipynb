```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter, MultipleLocator

# Assuming df is already loaded with your data

# Convert 'price' to numeric if it is not already
df['price'] = pd.to_numeric(df['price'], errors='coerce')

# Filter out rows where the number of bedrooms is 0
df = df[df['bedrooms'] != 0]

# Calculate average price by number of bedrooms
avg_price_by_bedrooms = df.groupby('bedrooms')['price'].mean()

# Find the index of the highest number of bedrooms
highest_bedrooms = avg_price_by_bedrooms.idxmax()

# Function to format the y-axis labels in thousands
def thousands_formatter(x, pos):
    return f'{int(x / 1000)}K'

# Create the bar chart
plt.figure(figsize=(10, 6))
colors = ['skyblue' if x != highest_bedrooms else 'darkblue' for x in avg_price_by_bedrooms.index]
avg_price_by_bedrooms.plot(kind='bar', color=colors, edgecolor='black')

plt.title('Average House Price by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average Price')
plt.xticks(rotation=0)  # Ensure x-ticks are horizontal

# Set y-axis major ticks and formatter
ax = plt.gca()
ax.yaxis.set_major_locator(MultipleLocator(200000))
ax.yaxis.set_major_formatter(FuncFormatter(thousands_formatter))

plt.show()
```
